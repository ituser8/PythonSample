```python
import numpy as np
import matplotlib.pyplot as plt
from torch import nn,optim
from torch.autograd import Variable
import torch
```

```python
#original dataset
x_data=np.random.rand(100)
print(x_data)
noise=np.random.normal(0,0.01,x_data.shape)
y_data=x_data*0.1+0.2+noise
plt.scatter(x_data,y_data)
plt.show()
```

```python
#reshape dimension dataset and create tensor
inputs=Variable(torch.FloatTensor(x_data.reshape(-1,1)))
target=Variable(torch.FloatTensor(y_data.reshape(-1,1)))
print(inputs)
print(target)
```

```python
#architement neural network
class LinearRegression(nn.Module):
    #initialize
    def __init__(self):  
        super(LinearRegression,self).__init__()
        self.fc=nn.Linear(1, 1)
        
    #define network calculation
    def forward(self,x):
        out=self.fc(x)
        return out
        
#define module
model=LinearRegression()
#define loss function
mse_loss=nn.MSELoss()
#define optimizer and learning rate
optimizer =optim.SGD(model.parameters(), lr=0.01)

#list parameter value in function
for name, parameters in model.named_parameters():
    print('name:{},param{}'.format(name, parameters))
```

```python
#calculate    
for i in range(3001):
    out=model(inputs)
    #calculate Loss
    loss=mse_loss(out,target)
    #initial gradient
    optimizer.zero_grad()
    #calculate gradient
    loss.backward()
    #weight
    optimizer.step()
    if i%100==0:
        print(i,loss.item())

#plot
y_pred=model(inputs)
plt.scatter(x_data, y_data)
plt.plot(x_data,y_pred.data.numpy(),'r-',lw=3)
plt.show()
```
